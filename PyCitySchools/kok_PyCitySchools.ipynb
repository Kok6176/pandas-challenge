```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data)
student_data_df = pd.read_csv(student_data)
```

```python
# Combine the 2 dataframes into a single dataset.  
school_merge_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_merge_df.head()
```

```python
# Calculate the total number of schools
total_schools = len(school_merge_df["school_name"].unique())
total_schools
```

```python
# Calculate the total number of students
total_students = len(school_merge_df["Student ID"])
total_students
```

```python
# Calculate the total budget
total_budget = (school_merge_df["budget"].unique()).sum()
total_budget
```

```python
# Calculate the average math score
total_aver_ms = school_merge_df["math_score"].mean()
total_aver_ms
```

```python
# Calculate the average reading score
total_aver_rs = school_merge_df["reading_score"].mean()
total_aver_rs
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
per_pass_ms = (school_merge_df[school_merge_df["math_score"] >= 70].math_score.count())/total_students *100
per_pass_ms
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
per_pass_rs = (school_merge_df[school_merge_df["reading_score"] >= 70].reading_score.count())/total_students *100
per_pass_rs
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
pass_ms = school_merge_df[school_merge_df["math_score"] >= 70]  
pass_mrs = (pass_ms[pass_ms["reading_score"] >= 70])
per_pass_mrs = pass_mrs.student_name.count() / total_students *100
# pass_ms
# pass_mrs
per_pass_mrs
```

```python
# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({"Total Schools" : [total_schools],
                                   "Total Students" : [total_students],
                                   "Total Budget" : [total_budget],
                                   "Average Math Score" : [total_aver_ms],
                                   "Average Reading Score" : [total_aver_rs],
                                   "% Passing Math" : [per_pass_ms],
                                   "% Passing Reading" : [per_pass_rs],
                                   "% Overall Passing" : [per_pass_mrs]})
district_summary_df
```

```python
# Optional: give the displayed data cleaner formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map('{:,}'.format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map('${:,.2f}'.format)
district_summary_df
```

<h1> Math Scores by Grade </h1>
<h4> Steps </h4>
<h6> 
    <ul> Create a table that lists the average math Score for students of each grade level (9th, 10th, 11th, 12th) at each school. </ul>
    <ul> define a function to take Grade as Input parameter and return a Dataframe having columns for school names and average math score for all students in that grade in that school.</ul>    
    <ul> Merge the Dataframes for each grade into one Summary Dataframe and set the index as School Name. </ul>
   
        
</h6>

```python
def average_math_by_grade(grade):
    sch_summary_msgrade = school_merge_df.loc[school_merge_df.grade == grade].groupby(['school_name'])
    sch_names_msgrade = school_merge_df.school_name.sort_values().unique()
    sch_avg_math_score = list(sch_summary_msgrade.math_score.mean())
  
   
    # Compile all calculations into dataframe
    average_ms_gr = pd.Series(sch_avg_math_score)
    average_ms_sn = pd.Series(sch_names_msgrade)
    
    average_math_df = pd.DataFrame({"school_names":average_ms_sn,f"{grade}":average_ms_gr})
    average_math_df = average_math_df.reset_index(drop=True)


   
    return average_math_df
```

```python
grade_9 = average_math_by_grade('9th')
grade_10 = average_math_by_grade('10th')
grade_11 = average_math_by_grade('11th')
grade_12 = average_math_by_grade('12th')
avg_math_score_by_grade = pd.merge(grade_9,grade_10,how='inner',suffixes=('',''))
avg_math_score_by_grade = pd.merge(avg_math_score_by_grade,grade_11,how='inner',suffixes=('',''))
avg_math_score_by_grade = pd.merge(avg_math_score_by_grade,grade_12,how='inner',suffixes=('',''))
avg_math_score_by_grade = avg_math_score_by_grade.set_index(school_names,drop=True)
avg_math_score_by_grade = avg_math_score_by_grade.loc[:, avg_math_score_by_grade.columns!='school_names']
avg_math_score_by_grade
```
